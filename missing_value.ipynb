# **Missing values**

```python
url='https://raw.githubusercontent.com/krishnaik06/Feature-Engineering-Live-sessions/master/titanic.csv'
```

```python
titanic_df=pd.read_csv(url)
```

```python
titanic_df.head()
```

```python
 titanic_df.isnull().sum()
```

```python
titanic_df[titanic_df['Embarked'].isnull()]# missing comepletly at random
```

```python
#missing data not at random
titanic_df['cabin_null']=np.where(titanic_df['Cabin'].isnull(),1,0)
#when there is null replace it with 1 and if there is not null replace it with 0
##find the percentage of null values
titanic_df['cabin_null'].mean()
```

```python
titanic_df.groupby(['Survived'])['cabin_null'].mean()
```

missing at random

```python
df=titanic_df[['Age','Fare','Survived']]
```

```python
df.head()
```

```python
def impute_na(df,variable,median):
  df[variable+'_median']=df[variable].fillna(median)
```

```python
median=df.Age.median()
median
```

```python
impute_na(df,'Age',median)
df.head()
```
